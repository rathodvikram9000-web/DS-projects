```python

def main():
    t_no = int(input("Enter Table Number: "))
    # cust_id = int(input("Enter Customer ID: "))
    m_id = int(input("Enter Menu ID: "))


    cur.execute(f"SELECT item_name, price FROM menu WHERE id = {m_id}")
    price1 = cur.fetchone()
    db.commit()

    if not price1:
        print("Invalid menu_id")
        return 

    item_name, price = price1
    qunt = int(input(f"Enter quantity of {item_name}: "))
    total_price = price * qunt

    insert_query1 = """
        INSERT INTO order_details (table_no, menu_id, quant, total)
        VALUES (%s, %s, %s, %s)
    """ 
    values = (t_no, m_id, qunt, total_price)

    cur.execute(insert_query1, values)
    db.commit()

    return t_no

while True:
    ask = input("Do you want to take order? (yes / stop): ").lower()

    if ask == "stop":
        print("Order taking stopped.")
        break

    t_no = main()   # take order for each table reques


# main()
```
